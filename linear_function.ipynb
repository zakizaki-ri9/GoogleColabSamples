```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.arange(0, 10, 0.1)
print(x)
```

```
# 1次関数の描画
# y = 2x + 1
y = 2 * x + 1
print(y)
plt.plot(x, y)
plt.show()
```

```
# 2次関数の描画
# y = x^2 - 10x + 10
y_2 = x**2 - 10*x + 10
plt.plot(x, y_2)
plt.show()
```

```
# 3次関数の描画
# y = x^2 - 10x + 10
y_2 = x**2 - 10*x + 10
# y = x^3 - 10x^2 - 10x + 10
y_3 = x**3 - 10*x**2 - 10*x + 10
plt.plot(x, y_2)
plt.plot(x, y_3)
plt.show()
```
